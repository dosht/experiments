Load

```python
from sklearn.datasets import load_iris
data = load_iris()
X = data.data
y = data.target
```

Preprocessing

```python
from sklearn.preprocessing import normalize
X_normailzed = normalize(X, norm='l2', axis=1, copy=True)
```

Constructing feedforward network

```python
from pybrain.structure import GaussianLayer, LinearLayer, SigmoidLayer, TanhLayer
from ann_helper import construct_ff_network
n = construct_ff_network(
    LinearLayer(4),
    [
        TanhLayer(6),
        TanhLayer(6),
    ],
    LinearLayer(3))
```

Fitting and simple validation

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_normailzed, y, test_size=0.3, random_state=0)

from rbf_nets import fit, create_dataset
ds = create_dataset(X_normailzed, y)
fit(n, ds, num_iterations=100)
```

```python
from ann_helper import predict, score
print score(n, X_train, y_train)
print score(n, X_test, y_test)
```

## Plotting maximum neuron responses

Plotting 2D data

```python
from itertools import cycle
import pylab as pl
def plot_2d(X, y, labels):
    colors = cycle("rgbcykw")
    pl.figure()
    labels_ids = range(len(labels))
    for i, c, l in zip(labels_ids, colors, labels):
        pl.scatter(X[y == i, 0], X[y == i, 1], c=c, label=l)
    
    pl.legend()
```

Reducing dimentions into 2D

```python
from sklearn.decomposition import PCA
pca = PCA(whiten=True, n_components=2).fit(X)
X_pca = pca.transform(X)
```

```python
plot_2d(X_pca, y, data.target_names)
```

Train a network with one tanh hidden layer

```python
n2 = construct_ff_network(LinearLayer(2), [TanhLayer(5)], LinearLayer(3))
ds = create_dataset(X_pca, y)
fit(n2, ds)
```

```python
score(n2, X_pca, y)
```

### Plotting neuron responses

```python
from itertools import cycle
import pylab as pl
def plot_neuron_responses(nn, X, y, labels):
    pl.figure(figsize=(10, 8))
    labels_ids = range(len(labels))
    markers = '*ox'
    colors = 'rgbcykw'
    for i, m, l in zip(labels_ids, markers, labels):
        for xi in X[y == i]:
            nn.activate(xi)
            c = colors[nn.modulesSorted[1].outputbuffer.argmax()]
            pl.scatter(xi[0], xi[1], c=c, marker=m, s=50)
```

```python
plot_neuron_responses(n2, X_pca, y, data.target_names)
```
