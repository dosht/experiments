```python
%pylab inline

# SVM Example using dataset: iris
```

```python
# Load
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
targets_names = iris.target_names
```

```python
# Plotting
from itertools import cycle
import pylab as pl
def plot_2d(X, y, labels):
    colors = cycle("rgbcykw")
    pl.figure()
    labels_ids = range(len(labels))
    for i, c, l in zip(labels_ids, colors, labels):
        pl.scatter(X[y == i, 0], X[y == i, 1], c=c, label=l)
    
    pl.legend()

# Plot all compinations of features
plot_2d(X[:, [0,1]], y, targets_names)
plot_2d(X[:, [0,2]], y, targets_names)
plot_2d(X[:, [0,3]], y, targets_names)
plot_2d(X[:, [1,2]], y, targets_names)
plot_2d(X[:, [1,3]], y, targets_names)
plot_2d(X[:, [2,3]], y, targets_names)
```

```python
# Deminsionality reduction
from sklearn.decomposition import PCA
pca = PCA(whiten=True, n_components=2).fit(X)
X_pca = pca.transform(X)
print "Dimantions before reduction: %s" % X.shape[1]
print "Dimantions after reduction: %s" % X_pca.shape[1]
```

```python
# Plotting after deminsionality reduction
pylab.figure()
plot_2d(X_pca, y, targets_names)
```

```python
# SVM Model
from sklearn.svm import LinearSVC
clf = LinearSVC('l2')
clf.fit(X, y)

clf.predict([[5., 3.6, 1.3, 0.3]])
```

```python
# Validation
validation = [i for i in range(X.shape[0]) if i % 4 == 0]
train = [i for i in range(X.shape[0]) if i % 4]

X_train, y_train = X[train, :], y[train]
X_validation, y_validation = X[validation, :], y[validation]

clf2 = LinearSVC('l2')
clf2.fit(X_train, y_train)
predictions2 = clf2.predict(X_validation)
accuracy2 = float((predictions2 == y_validation).sum()) / len(validation)
print accuracy2

clf1 = LinearSVC('l1', dual=False)
clf1.fit(X_train, y_train)
predictions1 = clf1.predict(X_validation)
accuracy1 = float((predictions1 == y_validation).sum()) / len(validation)
print accuracy1
```

```python
# K-Means example
from sklearn.cluster import KMeans
from numpy.random import RandomState
rng = RandomState(42)
kmeans = KMeans(k=3, random_state=rng)

kmeans.fit(X_pca)
plot_2d(X_pca, kmeans.labels_,  ['c1', 'c2', 'c3'])
```
